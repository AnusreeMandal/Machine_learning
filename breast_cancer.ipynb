## Define training and evaluation function

### Subtask:
Define a function to train and evaluate the Logistic Regression model with different solvers and penalties.


**Reasoning**:
Define a function to train and evaluate the Logistic Regression model with different solvers and penalties, handling potential errors and returning the model, accuracy, coefficients, and intercept.



```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

def train_evaluate_logistic_regression(X_train, y_train, X_val, y_val, solver, penalty):
    """
    Trains and evaluates a Logistic Regression model with specified solver and penalty.

    Args:
        X_train (pd.DataFrame): Training features.
        y_train (pd.Series): Training target.
        X_val (pd.DataFrame): Validation features.
        y_val (pd.Series): Validation target.
        solver (str): The solver to use ('newton-cg', 'lbfgs', 'liblinear').
        penalty (str): The penalty to use ('l1', 'l2', 'none').

    Returns:
        tuple: A tuple containing:
            - model (LogisticRegression): The trained Logistic Regression model.
            - accuracy (float): The accuracy on the validation set.
            - coefficients (np.ndarray): The coefficients of the trained model.
            - intercept (np.ndarray): The intercept of the trained model.
            Returns None for all if an error occurs.
    """
    try:
        # Instantiate the model with the specified solver and penalty
        # Set max_iter to a reasonable number to ensure convergence
        # For 'liblinear', only 'l1' and 'l2' penalties are supported
        # For 'newton-cg' and 'lbfgs', 'l2' and 'None' penalties are supported
        if solver == 'liblinear' and penalty == 'none':
            print(f"Warning: Penalty 'none' is not supported with solver 'liblinear'. Skipping.")
            return None, None, None, None
        elif solver in ['newton-cg', 'lbfgs'] and penalty == 'l1':
             print(f"Warning: Penalty 'l1' is not supported with solver '{solver}'. Skipping.")
             return None, None, None, None

        # Handle 'none' penalty correctly for newton-cg and lbfgs
        if penalty == 'none':
            model = LogisticRegression(solver=solver, penalty=None, max_iter=1000)
        else:
            model = LogisticRegression(solver=solver, penalty=penalty, max_iter=1000)


        # Train the model
        model.fit(X_train, y_train)

        # Predict on the validation data
        y_pred = model.predict(X_val)

        # Calculate accuracy
        accuracy = accuracy_score(y_val, y_pred)

        # Extract coefficients and intercept
        coefficients = model.coef_
        intercept = model.intercept_

        return model, accuracy, coefficients, intercept

    except Exception as e:
        print(f"Error training model with solver '{solver}' and penalty '{penalty}': {e}")
        return None, None, None, None

print("Function 'train_evaluate_logistic_regression' defined.")
```

## Train the model and create a table of coefficients

### Subtask:
Train the model with the training data and create a table for the coefficients for each combination of solver and penalty.

**Reasoning**:
Iterate through the specified solvers and penalties, train the Logistic Regression model using the training data with the `train_evaluate_logistic_regression` function, and collect the coefficients and intercept for each successful run. Then, store the coefficients in a dictionary, create a DataFrame from the dictionary, transpose it, and display the resulting table of coefficients.

```python
import pandas as pd

solvers = ['newton-cg', 'lbfgs', 'liblinear']
penalties = ['l1', 'l2', 'none']

initial_coefficients = {}

for solver in solvers:
    for penalty in penalties:
        print(f"Training with solver: {solver}, penalty: {penalty}")
        model, accuracy, coefficients, intercept = train_evaluate_logistic_regression(
            X_train, y_train, X_val, y_val, solver, penalty
        )

        if model is not None:
            # Store coefficients and intercept
            initial_coefficients[f'{solver}_{penalty}'] = coefficients.flatten().tolist() + [intercept[0]]

# Add a row for the intercept to the feature names
feature_names_with_intercept = list(X_train.columns) + ['Intercept']

# Create a DataFrame from the initial coefficients
initial_coefficients_df = pd.DataFrame(initial_coefficients, index=feature_names_with_intercept)

# Display the table of initial coefficients
print("\nInitial Coefficients Table:")
display(initial_coefficients_df)
```

```python
updated_coefficients = {}

for solver in solvers:
    for penalty in penalties:
        print(f"Fine-tuning with solver: {solver}, penalty: {penalty}")
        # Retrain the model with training and validation data combined for fine-tuning
        # In a real fine-tuning scenario, you might adjust hyperparameters based on validation performance.
        # For this task, we'll simply retrain on the combined data.
        X_train_val = pd.concat([X_train, X_val])
        y_train_val = pd.concat([y_train, y_val])

        model, accuracy, coefficients, intercept = train_evaluate_logistic_regression(
            X_train_val, y_train_val, X_test, y_test, solver, penalty
        )

        if model is not None:
            # Store updated coefficients and intercept
            updated_coefficients[f'{solver}_{penalty}'] = coefficients.flatten().tolist() + [intercept[0]]

# Create a DataFrame from the updated coefficients
updated_coefficients_df = pd.DataFrame(updated_coefficients, index=feature_names_with_intercept)

# Display the table of updated coefficients
print("\nUpdated Coefficients Table:")
display(updated_coefficients_df)
```

## Define a function

### Subtask:
Define a function to train and evaluate the Logistic Regression model with varying 'l1' penalty values and additional metrics.


**Reasoning**:
Define a function to train and evaluate the Logistic Regression model with varying 'l1' penalty values and additional metrics, handling potential errors and returning the model, evaluation metrics, coefficients, and intercept.



```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def train_evaluate_logistic_regression_l1(X_train, y_train, X_eval, y_eval, solver, C):
    """
    Trains and evaluates a Logistic Regression model with L1 penalty and specified solver and C.

    Args:
        X_train (pd.DataFrame): Training features.
        y_train (pd.Series): Training target.
        X_eval (pd.DataFrame): Evaluation features.
        y_eval (pd.Series): Evaluation target.
        solver (str): The solver to use.
        C (float): The inverse of regularization strength.

    Returns:
        tuple: A tuple containing:
            - model (LogisticRegression): The trained Logistic Regression model.
            - accuracy (float): The accuracy on the evaluation set.
            - precision (float): The precision on the evaluation set.
            - recall (float): The recall on the evaluation set.
            - f1 (float): The F1-score on the evaluation set.
            - coefficients (np.ndarray): The coefficients of the trained model.
            - intercept (np.ndarray): The intercept of the trained model.
            Returns None for all if an error occurs or solver/penalty combination is unsupported.
    """
    try:
        # Instantiate the model with the specified solver, L1 penalty, and C
        # Handle unsupported solver/penalty combinations
        if solver not in ['liblinear', 'saga']: # Solvers that support L1 penalty in scikit-learn
             print(f"Warning: Solver '{solver}' does not support 'l1' penalty. Skipping.")
             return None, None, None, None, None, None, None

        model = LogisticRegression(solver=solver, penalty='l1', C=C, max_iter=1000, random_state=5)

        # Train the model
        model.fit(X_train, y_train)

        # Predict on the evaluation data
        y_pred = model.predict(X_eval)

        # Calculate evaluation metrics
        accuracy = accuracy_score(y_eval, y_pred)
        precision = precision_score(y_eval, y_pred)
        recall = recall_score(y_eval, y_pred)
        f1 = f1_score(y_eval, y_pred)

        # Extract coefficients and intercept
        coefficients = model.coef_
        intercept = model.intercept_

        return model, accuracy, precision, recall, f1, coefficients, intercept

    except ValueError as ve:
        print(f"ValueError training model with solver '{solver}' and C='{C}': {ve}")
        return None, None, None, None, None, None, None
    except Exception as e:
        print(f"Error training model with solver '{solver}' and C='{C}': {e}")
        return None, None, None, None, None, None, None

print("Function 'train_evaluate_logistic_regression_l1' defined.")
```

Train the model with varying 'l1' penalties on the training data and evaluate on the validation data.

```python
solvers_l1 = ['liblinear', 'saga']
c_values = [0.1, 0.25, 0.75, 0.9]

l1_results = {}

for solver in solvers_l1:
    for c in c_values:
        print(f"Training with solver: {solver}, C: {c}")
        model, accuracy, precision, recall, f1, coefficients, intercept = train_evaluate_logistic_regression_l1(
            X_train, y_train, X_val, y_val, solver, c
        )

        if model is not None:
            l1_results[f'{solver}_C={c}'] = {
                'Accuracy': accuracy,
                'Precision': precision,
                'Recall': recall,
                'F1-score': f1,
                'Coefficients': coefficients.flatten().tolist() + [intercept[0]]
            }

# Separate metrics and coefficients
l1_metrics = {key: {k: v for k, v in val.items() if k != 'Coefficients'} for key, val in l1_results.items()}
l1_coefficients = {key: val['Coefficients'] for key, val in l1_results.items()}

# Add a row for the intercept to the feature names
feature_names_with_intercept = list(X_train.columns) + ['Intercept']

# Create DataFrames
l1_metrics_df = pd.DataFrame.from_dict(l1_metrics, orient='index')
l1_coefficients_df = pd.DataFrame(l1_coefficients, index=feature_names_with_intercept)

# Display the tables
print("\nL1 Penalty Evaluation Metrics on Validation Set:")
display(l1_metrics_df)

print("\nL1 Penalty Coefficients Table (on Training Data):")
display(l1_coefficients_df)
```
